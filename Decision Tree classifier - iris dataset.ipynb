## Task 4 - Decision Tree classifier-Iris dataset

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
# Importing the data
path="C:/Users/thakr/Downloads/Iris (1).csv"
data=pd.read_csv(path)
data.head()
```

```python
data.describe()
```

```python
#Plotting pairwise relationship of dataset
sns.pairplot(data,hue="Species")
```

```python
cols=["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
cols
```

```python
# preparing the data
X=data[cols]
y=data.Species
```

```python
# Training the algorithm
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)

X_train
```

```python
y_train
```

```python
# Defining the descision tree algorithm
clf=DecisionTreeClassifier()
clf
```

```python
clf = clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn import tree

plt.figure(figsize=(40,30))
tree.plot_tree(clf.fit(X,y),feature_names=cols,filled=True,precision=4,
              proportion=True,rounded=True,fontsize=25)
plt.show()
```
